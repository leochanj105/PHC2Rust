```python
import numpy as np
import matplotlib.pyplot as plt

# Data (replace with your actual values, in percentage form e.g. 0-100)
groups = ['s1', 's2', 's3', 's4', 's5']
values_a = [70, 75, 80, 85, 90]   # first bar in each group
values_b = [65, 78, 82, 88, 92]   # second bar in each group

x = np.arange(len(groups))  # group positions
width = 0.35                # width of each bar

plt.figure()

# Plot bars
plt.bar(x - width/2, values_a, width, label='A')
plt.bar(x + width/2, values_b, width, label='B')

# Axes and labels
plt.xticks(x, groups)
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # assuming percentage scale

# Optional: grid + legend
plt.grid(axis='y')
plt.legend()

plt.tight_layout()
plt.show()
```
